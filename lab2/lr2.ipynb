```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

Визуализация распределений и рассчет мат. ожидания, дисперсии, ассиметрии, эксцесс моментов. Квантили уровня 0,05 и 0,95; 2,5%-ную точку  для найденной случайной величины. Сравнительная диаграмма найденного распределения и распределения по нормальному закону.

1. Анализ данных о ценах на авокадо (источник: https://www.kaggle.com/datasets/neuromusic/avocado-prices)

Чтение данных и визуализация распеределения

```python
data = pd.read_csv("avocado.csv")
data
# Извлечение столбца со средней ценой
prices = data["AveragePrice"].dropna()  # Убираем пропуски

# Построение гистограммы
plt.hist(prices, bins=20, density=True, alpha=0.6, color="g", label="Средняя цена авокадо")

# Оценка параметров нормального распределения
mu_1, std_1 = prices.mean(), prices.std()

# Создание кривой нормального распределения
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_1, std_1)

# Наложение кривой нормального распределения
plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

# Оформление графика
plt.title("Распределение средней цены авокадо")
plt.xlabel("Цена (в долларах)")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

Расчет характеристик распределения, близкого к нормальному

```python
# 1. Математическое ожидание (среднее значение)
mean_value_1 = prices.mean()
print(f"Математическое ожидание: {mean_value_1:.3f} долларов")

# 2. Дисперсия
variance_1 = prices.var()
print(f"Дисперсия: {variance_1:.3f} долларов")

# 3. Асимметрия
skewness_1 = prices.skew()
print(f"Асимметрия: {skewness_1:.3f}")

# 4. Эксцесс
kurtosis_1 = prices.kurtosis()
print(f"Эксцесс: {kurtosis_1:.3f}")

# 5. Квантили уровня 0,05 и 0,95
quantile_05_1 = prices.quantile(0.05)
quantile_95_1 = prices.quantile(0.95)
print(f"Квантиль 0,05: {quantile_05_1:.3f} долларов")
print(f"Квантиль 0,95: {quantile_95_1:.3f} долларов")

# 2,5%-ная точка
quantile_025_1 = prices.quantile(0.025)
print(f"2,5%-ная точка: {quantile_025_1:.3f} долларов")
```

- **Математическое ожидание:** Средняя цена на авокадо составляет 1.406 долларов, что можно считать типичной ценой в данной выборке.

- **Дисперсия:** Дисперсия 0.162 указывает на умеренную вариативность цен на авокадо, т.е. цены относительно стабильно варьируются вокруг среднего значения.

- **Асимметрия:** Асимметрия 0.580 говорит о том, что распределение цен немного сдвинуто вправо, то есть высокие цены встречаются чуть чаще, чем низкие.

- **Эксцесс:** Эксцесс 0.325 указывает на слегка "заостренное" распределение, что говорит о наличии более выраженного "пика" вокруг среднего, чем в случае нормального распределения.

- **Квантиль 0.05:** Лишь 5% цен находятся ниже 0.830 долларов, что отражает нижнюю границу стоимости для большинства наблюдений.

- **Квантиль 0.95:** Большинство (95%) цен ниже 2.110 долларов, и только 5% превышают этот уровень, что показывает редкость высоких цен.

- **2,5%-ная точка:** Только 2,5% цен на авокадо находятся ниже 0.750 долларов, что подчеркивает редкость очень низких цен в выборке.

2. Анализ данных о росте людей (источник: https://www.kaggle.com/datasets/krishnaraj30/weight-and-height-data-outlier-detection)

```python
data = pd.read_csv("weight-height.csv")
data
# Извлечение столбца со средней ценой
prices = data["Height"].dropna()  # Убираем пропуски

# Построение гистограммы
plt.hist(prices, bins=20, density=True, alpha=0.6, color="b", label="Рост")

# Оценка параметров нормального распределения
mu_1, std_1 = prices.mean(), prices.std()

# Создание кривой нормального распределения
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_1, std_1)

# Наложение кривой нормального распределения
plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

# Оформление графика
plt.title("Распределение роста людей")
plt.xlabel("Рост (в дюймах)")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

```python
# 1. Математическое ожидание (среднее значение)
mean_value_2 = prices.mean()
print(f"Математическое ожидание: {mean_value_2:.3f} дюймов")

# 2. Дисперсия
variance_2 = prices.var()
print(f"Дисперсия: {variance_2:.3f} дюймов")

# 3. Асимметрия
skewness_2 = prices.skew()
print(f"Асимметрия: {skewness_2:.3f}")

# 4. Эксцесс
kurtosis_2 = prices.kurtosis()
print(f"Эксцесс: {kurtosis_2:.3f}")

# 5. Квантили уровня 0,05 и 0,95
quantile_05_2 = prices.quantile(0.05)
quantile_95_2 = prices.quantile(0.95)
print(f"Квантиль 0,05: {quantile_05_2:.3f} дюймов")
print(f"Квантиль 0,95: {quantile_95_2:.3f} дюймов")

# 2,5%-ная точка
quantile_025_2 = prices.quantile(0.025)
print(f"2,5%-ная точка: {quantile_025_2:.3f} дюймов")
```

- **Математическое ожидание:** Средний рост в популяции составляет 66.368 дюймов, что можно считать типичным значением для этой выборки.

- **Дисперсия:** Дисперсия 14.803 указывает на значительную вариативность роста в популяции, то есть рост сильно колеблется относительно среднего.

- **Асимметрия:** Асимметрия 0.049 близка к нулю, что свидетельствует о симметричном распределении роста, без значительного перекоса в одну из сторон.

- **Эксцесс:** Эксцесс -0.474 указывает на более «плоское» распределение по сравнению с нормальным, что говорит о менее выраженных «пиках» около среднего значения.

- **Квантиль 0.05:** Только 5% людей имеют рост ниже 60.246 дюймов, что отражает низшую границу значений для большинства выборки.

- **Квантиль 0.95:** Большинство людей (95%) имеют рост ниже 72.617 дюймов, и только 5% превышают этот показатель, что помогает оценить верхнюю границу роста.

- **2,5%-ная точка:** Лишь 2,5% людей имеют рост ниже 59.264 дюймов, что подчеркивает редкость очень низких значений роста.
